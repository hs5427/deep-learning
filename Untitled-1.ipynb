```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import plotly.express as px
```

```python
def plot_distribution(distribution_name, **params):
    # x
    x = np.linspace(-10, 10, 1000)
    
    # PDF
    dist = getattr(stats, distribution_name)
    y = dist.pdf(x, **params)
    
    # 
    df = {
        'x': x,
        'y': y
    }
    
    # Plotly Express
    fig = px.line(df, x='x', y='y', title=f"{distribution_name.capitalize()} Distribution")
    
    # 
    fig.show()

# :  (Normal Distribution) 
plot_distribution('norm', loc=0, scale=1)

# :  (Gamma Distribution) 
plot_distribution('gamma', a=2, scale=2)
```

```python
import numpy as np
import plotly.graph_objects as go

# A
mu_A = 50
sigma_A = 5

# B
mu_B = 60
sigma_B = 10

# x, y
x = np.linspace(mu_A - 3*sigma_A, mu_A + 3*sigma_A, 100)
y = np.linspace(mu_B - 3*sigma_B, mu_B + 3*sigma_B, 100)

# 
X, Y = np.meshgrid(x, y)

# 2
Z = (1 / (2 * np.pi * sigma_A * sigma_B) *
     np.exp(-((X - mu_A)**2 / (2 * sigma_A**2) + (Y - mu_B)**2 / (2 * sigma_B**2))))

# Plotly3D
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# 
fig.update_layout(title="2D Normal Distribution of Dimensions A and B",
                  scene = dict(
                    xaxis_title='Dimension A',
                    yaxis_title='Dimension B',
                    zaxis_title='Probability Density'),
                  width=800, height=700)

# 
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# 
mu_A, sigma_A = 50, 5
mu_B, sigma_B = 60, 10

# 
rho = -0.8

# 
cov_matrix = [[sigma_A**2, rho * sigma_A * sigma_B],
              [rho * sigma_A * sigma_B, sigma_B**2]]

# 
x_A = np.linspace(mu_A - 3*sigma_A, mu_A + 3*sigma_A, 100)
x_B = np.linspace(mu_B - 3*sigma_B, mu_B + 3*sigma_B, 100)
X_A, X_B = np.meshgrid(x_A, x_B)

# 2
pos = np.empty(X_A.shape + (2,))
pos[:, :, 0] = X_A
pos[:, :, 1] = X_B
rv = stats.multivariate_normal([mu_A, mu_B], cov_matrix)
Z = rv.pdf(pos)

# Plotly3D
fig = go.Figure(data=[go.Surface(z=Z, x=X_A, y=X_B)])

# 
fig.update_layout(title="2D Correlated Normal Distribution of Dimensions A and B",
                  scene = dict(
                    xaxis_title='Dimension A',
                    yaxis_title='Dimension B',
                    zaxis_title='Probability Density'),
                  width=800, height=700)

# 
fig.show()
```

```python
import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

# 
mu = [50, 60, 70, 80]

# 
sigma_A, sigma_B, sigma_C, sigma_D = 5, 10, 7, 15

# 
rho_AD, rho_BD, rho_CD = -0.5, -0.4, -0.6

# 
cov_matrix = [
    [sigma_A**2, 0, 0, rho_AD * sigma_A * sigma_D],
    [0, sigma_B**2, 0, rho_BD * sigma_B * sigma_D],
    [0, 0, sigma_C**2, rho_CD * sigma_C * sigma_D],
    [rho_AD * sigma_A * sigma_D, rho_BD * sigma_B * sigma_D, rho_CD * sigma_C * sigma_D, sigma_D**2]
]

# 
rv = stats.multivariate_normal(mean=mu, cov=cov_matrix)

# AD
x_A = np.linspace(mu[0] - 3*sigma_A, mu[0] + 3*sigma_A, 100)
x_D = np.linspace(mu[3] - 3*sigma_D, mu[3] + 3*sigma_D, 100)

# 
X_A, X_D = np.meshgrid(x_A, x_D)

# 2BC
pos = np.empty(X_A.shape + (4,))
pos[:, :, 0] = X_A
pos[:, :, 1] = mu[1]  # B
pos[:, :, 2] = mu[2]  # C
pos[:, :, 3] = X_D
Z = rv.pdf(pos)

# Plotly3D
fig = go.Figure(data=[go.Surface(z=Z, x=X_A, y=X_D)])

# 
fig.update_layout(title="2D Marginal Distribution of Dimensions A and D",
                  scene = dict(
                    xaxis_title='Dimension A',
                    yaxis_title='Dimension D',
                    zaxis_title='Probability Density'),
                  width=800, height=700)

# 
fig.show()
```

```python
import numpy as np
import scipy.stats as stats
from scipy.integrate import nquad

# 
mu = [50, 60, 70, 80]

# 
sigma_A, sigma_B, sigma_C, sigma_D = 5, 10, 7, 15

# 
rho_AD, rho_BD, rho_CD = -0.5, -0.4, -0.6

# 
cov_matrix = [
    [sigma_A**2, 0, 0, rho_AD * sigma_A * sigma_D],
    [0, sigma_B**2, 0, rho_BD * sigma_B * sigma_D],
    [0, 0, sigma_C**2, rho_CD * sigma_C * sigma_D],
    [rho_AD * sigma_A * sigma_D, rho_BD * sigma_B * sigma_D, rho_CD * sigma_C * sigma_D, sigma_D**2]
]

# 
rv = stats.multivariate_normal(mean=mu, cov=cov_matrix)

#  (x_A, x_B, x_C, x_D)
x_A_range = (51.5, 52.5)
x_B_range = (61.5, 62.5)
x_C_range = (67.5, 68.5)
x_D_range = (74.5, 75.5)

# 
def integrand(x_A, x_B, x_C, x_D):
    return rv.pdf([x_A, x_B, x_C, x_D])

# 
prob, error = nquad(integrand, [x_A_range, x_B_range, x_C_range, x_D_range])

prob, error
```

```python
import numpy as np
import scipy.stats as stats

# 
mean_A, mean_B, mean_C, mean_D = 0, 0, 0, 0
std_A, std_B, std_C, std_D = 1, 1, 1, 1
correlation_DA = -0.5  # DA
correlation_DB = -0.5  # DB
correlation_DC = -0.5  # DC

# ABC
num_samples = 100000
samples_A = np.random.normal(mean_A, std_A, num_samples)
samples_B = np.random.normal(mean_B, std_B, num_samples)
samples_C = np.random.normal(mean_C, std_C, num_samples)

# D
cov_matrix = np.array([
    [std_A**2, correlation_DA * std_A * std_D],
    [correlation_DA * std_A * std_D, std_D**2]
])
mean_D = mean_A * correlation_DA  # 
samples_D = np.random.multivariate_normal([mean_A, mean_D], cov_matrix, num_samples)[:, 1]

# 
interval = 0.1
bins = np.arange(0, 4 + interval, interval)
hist_A, _ = np.histogram(samples_A, bins)
hist_B, _ = np.histogram(samples_B, bins)
hist_C, _ = np.histogram(samples_C, bins)
hist_D, _ = np.histogram(samples_D, bins)

# 
prob_A = hist_A / num_samples
prob_B = hist_B / num_samples
prob_C = hist_C / num_samples
prob_D = hist_D / num_samples

# 
def calculate_joint_probability(bin_idx_A, bin_idx_B, bin_idx_C):
    prob_A_bin = prob_A[bin_idx_A]
    prob_B_bin = prob_B[bin_idx_B]
    prob_C_bin = prob_C[bin_idx_C]
    
    # D
    # ABC
    prob_D_bin = prob_D[bin_idx_A]
    joint_prob = prob_A_bin * prob_B_bin * prob_C_bin * prob_D_bin
    return joint_prob

# ABC [0, 0.1) 
bin_idx_A = 0
bin_idx_B = 0
bin_idx_C = 0
joint_probability = calculate_joint_probability(bin_idx_A, bin_idx_B, bin_idx_C)
print(f"ABC [0, 0.1) D: {joint_probability}")
```
